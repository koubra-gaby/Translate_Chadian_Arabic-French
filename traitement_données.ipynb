```python
import PyPDF2 
# Ouvrir un fichier PDF
inputFile = "/home/koubra/Documents/Corpus A_T/lexique_arb.pdf"
pdf = open(inputFile, "rb")
pdf_reader = PyPDF2.PdfReader(pdf)
#page = pdf_reader.pages[4]
parts = []


def visitor_body(text, cm, tm, font_dict, font_size):
    y = cm[2]
    if  y == text.isnumeric():
        parts.append(text)

for i in range(4, len(pdf_reader.pages)) :
    page = pdf_reader.pages[i]
    # Extraire le texte de la page
    page.extract_text(visitor_text=visitor_body)
    text_body = "".join(parts)
text_body = [i for i in text_body.split() if i.isascii()]
#text_body = text_body.replace("?", "\n")
with open('abu-samah.txt', 'a+') as file :
        for i in text_body:
            if i.isnumeric():
                i = i.replace(i, "\n")
            file.write(i)
            file.write("\n")

print(text_body)
```
